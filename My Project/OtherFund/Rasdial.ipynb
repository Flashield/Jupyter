```python
import os,time,random,csv
#提前建一个宽带拨号连接，连接名称填入dialname
dialname = 'Dial'
```

```python
filename = 'Dial.csv'
with open(filename) as f:
    reader = csv.reader(f)
    account = list(reader)
    #print(account)
```

```python
class pppoe:
    def __init__(self):
        #self.name = g_adsl_account["name"]
        #self.username = list[0]
        #self.password = list[1]
        pass
        
    def connect(str,list):
        cmd_str = "rasdial {} {} {}".format(str, list[0], list[1])
        #os.system(cmd_str)
        print(cmd_str)
        time.sleep(5)
        
    def disconnect(dialname):
        cmd_str = "rasdial {} /disconnect".format(dialname)
        #os.system(cmd_str)
        print(cmd_str)
        time.sleep(3)
    
    def get_flow():
        pack_cnt = random.randint(600,1000)
        pack_len = random.randint(1500,5000)
        cmd_str = "ping www.sina.com -n {} -l {}".format(pack_cnt,pack_len)
        #os.system(cmd_str)
        print(cmd_str)
        
```

```python
while True:
    for i in range(len(account)):
        pppoe.connect(dialname,account[i])
        pppoe.get_flow()
        pppoe.disconnect(dialname)
```
