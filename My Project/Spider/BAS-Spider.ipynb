```python
%env DISPLAY=192.168.245.1:0
#%env DISPLAY=211.139.119.233:0
from selenium import webdriver
#from selenium.webdriver.common.by import By
#from selenium.webdriver.support.ui import WebDriverWait
#from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
from bs4 import BeautifulSoup
```

```python
chrome_options = webdriver.chrome.options.Options()
#chrome_options.add_argument("--window-size=1024x768")
#chrome_options.add_argument("--headless")
```

```python
browser = webdriver.Chrome(options=chrome_options)
#browser = webdriver.Firefox()
```

```python
#登陆4A
browser.get('http://new4a.js.cmcc/uac/web3/jsp/login/login.jsp')
browser.implicitly_wait(10)
user_name=input('Input your 4A UserName：')
browser.find_element_by_id('loginName').send_keys(user_name)
user_password=input('Input your 4A Password：')
#密码登陆
Ele_Password = browser.find_element_by_id('passwd_input_placeholder')
Act_Password = ActionChains(browser).move_to_element(Ele_Password)
Act_Password.click()
Act_Password.send_keys(user_password)
Act_Password.perform()
browser.find_element_by_id('login_btn').click()
#验证码登陆
user_Verify_Code=input('Input your 4A SMS Verify Code：')
browser.find_element_by_id('smKey').send_keys(user_Verify_Code)
browser.find_element_by_id('login_btn').click()
```

```python
#登陆BAS
data_frame = browser.find_element_by_id('title3tab') # 定位frame元素
browser.switch_to.frame(data_frame)
#鼠标悬停
BAS_Menu=browser.find_element_by_xpath('//*[@id="appLogin_div0"]')
ActionChains(browser).move_to_element(BAS_Menu).perform()
data_frame = browser.find_element_by_xpath('//*[@id="appLoginChoose_div0"]/iframe') # 定位frame元素
browser.switch_to.frame(data_frame)
browser.find_element_by_id('appAccess').click()

current_windows = browser.window_handles
browser.switch_to.window(current_windows[1])
#进入BAS
browser.get(browser.current_url)
```
