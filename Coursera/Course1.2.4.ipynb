```python
def isprime(x):
    import math
    if x == 1:
        return False
    k = int(math.sqrt(x))
    for j in range(2, k+1):
         if x % j == 0:
             return False
    else:
        return True
```

```python
#-*-coding:utf-8-*-
def isPrime(n):
    i = 2
    while i < n // 2 + 1:
        if n % i == 0:
            return False
        i += 1
    return True

m = 4
count = 0
while m < 1000:
    i = 2
    while i < m:
        if isPrime(i) and isPrime(m-i):
            count += 1
            if count % 6 != 0:
                print("{}={}+{}".format(m,i,m-i),end=" ")
            else:
                print("{}={}+{}".format(m,i,m-i))
            m += 2
            #break
        else:
            i += 1
```
